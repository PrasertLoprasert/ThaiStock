```python
# ใช้ driver เก็บข้อมูลในตารางหุ้นแต่ละตัว ไว้ในตัวแปร all_stock_dict โดยมีข้อมูล 2 ตาราง

all_stock_dict = dict()
driver = webdriver.Chrome()

for st in stock['หลักทรัพย์']:
    url = 'https://www.set.or.th/th/market/product/stock/quote/'+ st +'/financial-statement/company-highlights'
    driver.get(url)

    stock_data = driver.page_source
    a_df = pd.read_html(stock_data)[0]
    b_df = pd.read_html(stock_data)[1]
    all_stock_dict[st] = [a_df, b_df]
```

```python
# เปลี่ยน index เป็นชื่อ column
for key in all_stock_dict.keys():
    
    all_stock_dict[key][0].set_index('งวดงบการเงิน ณ วันที่', inplace=True)   
    all_stock_dict[key][1].set_index('ค่าสถิติสำคัญ ณ วันที่', inplace=True)
```

```python
# ข้อมูลที่ต้องการมีดังนี้
# 1.Price
# 2.P/E
# 3.Dividend %
# 4. Cash / Total Debt
# 5. Avg.Profit(3 years) 

```

```python
all_stock_dict = dict()
driver = webdriver.Chrome()

for st in stock['หลักทรัพย์']:
    url = 'https://www.set.or.th/th/market/product/stock/quote/'+ st +'/financial-statement/latest/balance'
    driver.get(url)
    stock_data = driver.page_source
    a_df = pd.read_html(stock_data)[0]
    b_df = pd.read_html(stock_data)[1]
    all_stock_dict[st] = [a_df, b_df]
```

```python
for key in all_stock_dict.keys():
   
     all_stock_dict[key][0].set_index(0, inplace=True)   
     all_stock_dict[key][1].set_index(0, inplace=True)
```

```python
select_row = 'อัตราส่วนเงินปันผลตอบแทน (%)'
result_dict = dict()

for key in all_stock_dict.keys():
    try:
        money = float(all_stock_dict[key][0][1]['เงินสดและรายการเทียบเท่าเงินสด'])
        debt = float(all_stock_dict[key][1][1]['รวมหนี้สิน'])
        select_data = money / debt
        result_dict[key] = float(select_data)
    except KeyError:
        pass
    except ValueError:
        pass
```

```python
result_dict
```

```python
result_df = pd.DataFrame([result_dict])
result_df = result_df.transpose()
```

```python
result_df.columns = ['% Cash/Total Debt']
```

```python
result_df
```

```python
result_df = result_df.sort_values(by='% Cash/Total Debt', ascending=True)
result_df
```

```python
import matplotlib.pyplot as plt

threshold = 100000  # Set your threshold value as needed

# Filter the DataFrame to exclude values above the threshold
result_plot = result_df[result_df['% Cash/Total Debt'] <= threshold]

plt.figure(figsize=(20, 10))
plt.grid(True, linestyle='--', color='gray', alpha=0.5)
plt.barh(result_plot.index, result_plot['% Cash/Total Debt'])
plt.xlabel('Cash / Total Debt')
plt.ylabel('SET50')
plt.title(' Cash / Total Debt .SET50 , (Data Date: 2023-10-06)')
plt.show()
```
